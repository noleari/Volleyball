```python
import requests
from bs4 import BeautifulSoup

headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
```

```python
response = requests.get("https://stats.ncaa.org/rankings/national_ranking?academic_year=2024.0&division=1.0&ranking_period=81.0&sport_code=WVB&stat_seq=51.0&_=1702584265474", headers=headers)
response
```

```python
soup = BeautifulSoup(response.content)
```

```python
a = soup.find_all("a")
a
```

```python
team_id = {}
for element in a:
    if 'href' not in element.attrs: continue
    
    if "/team/" not in element.attrs["href"]: continue
    id = int(element.attrs["href"].split("/")[2][:-2])
    team = element.text.split("(")[0].strip()
    team_id[id] = team
```

```python
team_id
```

```python
with open("team_id.csv", "w") as f:
    f.write("team, id\n")
    for id, team in team_id.items():
        f.write(f"{team}, {id}\n")
```
